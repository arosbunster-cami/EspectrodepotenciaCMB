```python
Espectro = pd.read_csv('../Desktop/D_ell.csv')
D_ell = Espectro.iloc[:,1] 
_ell = Espectro.iloc[:,0] 
```

```python
#Resolución de 1'
```

```python
#M Tilde 
#Primero hay que obtener los valores de C_l

C_l=[];
for l in range(2,5050,1):
    C_l=(2 * math.pi* D_ell)/(l*(l+1))      

C_l2=np.zeros((16550))
C_l1=np.concatenate((C_l,C_l2))
```

```python
#Generar M tilde
MT1=[]
for lx in range(26,21600,36):
    mt1=[]
    for ly in range(26,21600,36):
        if math.sqrt(lx**2+ly**2)<=5049:
            m1=C_l[round(math.sqrt(lx**2+ly**2))]
        else:
            m1=0
        mt1.append(m1)
    MT1.append(mt1)
```

```python
#G Tilde

#N para una dimensión
mu,sigma=0,1   #defino los valores de mu y sigma para este espectro angular de potencias

N1 = np.random.normal(mu,sigma, size=(600,600))
```

```python
Gt1=(numpy.fft.fft2(N1))/(1/60/180*np.pi)
```

```python
# M 
M1=(numpy.fft.ifft2(MT1*Gt1))
```

```python
im=plt.imshow(np.real(M1),cmap=cm.RdBu_r)
plt.colorbar(label=r"Temperature$[\mu k]$");
#plt.clim(-300e-6,300e-6)
im.set_extent([0,10,0,10])
plt.xlabel('$\Theta_x(deg)$')
plt.ylabel('$\Theta_y(deg)$')
#plt.savefig('CMB1ma.pdf')
```

```python
plt.imshow(np.log(MT1))
im.set_extent([0,10,0,10])
plt.xlabel('$l_x$')
plt.ylabel('$l_y$')
plt.savefig('log1ma.pdf')
ax = plt.gca()
ax.set_ylim(ax.get_ylim()[::-1])
```
